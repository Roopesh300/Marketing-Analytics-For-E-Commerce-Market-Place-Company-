b. Understanding how many new customers acquired every month


```python
orders_cust = pd.merge(left = orders , right = customers , left_on = 'customer_id' , right_on = 'customer_id' , how = 'left')

# creating a year-month column to group by months

orders_cust['year_month'] = orders_cust['order_purchase_timestamp'].dt.to_period('M')

new_customers = orders_cust.groupby(['year_month', 'customer_unique_id']).nth(0).reset_index()

new_customers_count = new_customers.groupby('year_month').size().reset_index(name='new_customers_count')

new_customers_count
```

c. Understand the retention of customers on month on month basis

```python
orders_cust['cohort'] = orders_cust.groupby(by = 'customer_unique_id')['year_month'].transform(min)

cohort = orders_cust.groupby(by = ['year_month' , 'cohort']).agg({'customer_unique_id' : 'nunique'}).reset_index()

cohort['retention_rate'] = cohort['customer_unique_id'] / cohort.groupby('cohort')['customer_unique_id'].transform('first') * 100

retention_matrix = cohort.pivot_table(index='cohort', columns='year_month', values='retention_rate')

retention_matrix.fillna(0,inplace=True)

retention_matrix
```

d. How the revenues from existing/new customers on month on month basis

```python
# merge
orders_cust_payments = pd.merge(left = orders_cust , right = order_payments , left_on = 'order_id' , right_on = 'order_id' ,how = 'left')

# groupby
monthly_revenues = orders_cust_payments.groupby(['year_month', 'customer_unique_id'])['payment_value'].sum().reset_index()

# roupby with min dates
monthly_revenues['first_purchase_month'] = monthly_revenues.groupby('customer_unique_id')['year_month'].transform('min')

monthly_revenues['customer_type'] = 'Existing'
monthly_revenues.loc[monthly_revenues['year_month'] == monthly_revenues['first_purchase_month'], 'customer_type'] = 'New'

total_revenue = monthly_revenues.groupby(by = ['year_month' , 'customer_type']).agg(revenue = ('payment_value', 'sum')).reset_index()

pivot_table = total_revenue.pivot_table(index='year_month', columns='customer_type', values='revenue', aggfunc='sum')

pivot_table.fillna(0,inplace=True)

pivot_table


```
